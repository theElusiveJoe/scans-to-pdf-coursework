```python
%load_ext autoreload
%autoreload 2

import os
import config
os.chdir(config.PROJECT_ROOT_PATH)

import random
import torch
from torchvision.transforms.functional import to_pil_image

import numpy as np

import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from pylab import rcParams
rcParams['figure.figsize'] = 3,4
%matplotlib inline

from src.faster_rcnn.model import create_model
from src.faster_rcnn.trainer import Trainer
from src.faster_rcnn.datasets import MyDatasetValid
```

```python
colors = {
    1: 'red',
    2: 'yellow',
    3: 'magenta',
    4: 'blue',
    5: 'green',
}

def draw_img(x, bboxes=None, labels=None):
    img = to_pil_image(x)
    pix = np.array(img.getdata()).reshape(img.size[0], img.size[1], 3)
    plt.subplot()
    plt.imshow(pix)
    if bboxes is not None and labels is not None:
        for bbox, label in zip(bboxes, labels):
            x1,y1,x2,y2 = bbox
            h = (y2-y1)
            w = (x2-x1)
            plt.gca().add_patch(Rectangle((x1,y1),w,h,
                                          edgecolor=colors[int(label.item())],
                                          facecolor='none',
                                          lw=1))
    plt.show()
```
